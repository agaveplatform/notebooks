```bash
# we will make some test data to test out the regex support
# a meaninful naming scheme is important here
rm -rf rangetestdata
mkdir -p rangetestdata
for i in {1..3}; do 
    echo "test data file $i" > rangetestdata/file$i.txt
done
for i in foo baz bar; do 
    echo "test data file $i" > rangetestdata/$i.txt
done
```

```bash
# upload the data to a test folder
files-mkdir -N rangetest $AGAVE_USERNAME
for i in {1..3}; do 
    files-upload -V -F ./rangetestdata $AGAVE_USERNAME/rangetest;
done
```

```bash
# we should now be able to download the data into a flattened directory space using the 
# following syntax
mkdir rangetestdownload
cd rangetestdownload
files-get -V $AGAVE_USERNAME/rangetest/rangetestdata/file[1-3].txt
```

```bash
# we should also be able to download files into a flattened directory space using a bracket list notation
files-get -V "$AGAVE_USERNAME/rangetest/rangetestdata/{foo,bar,baz}.txt"
```
